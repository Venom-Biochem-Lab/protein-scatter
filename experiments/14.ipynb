```python
import pandas as pd
```

```python
df = pd.read_parquet("./datasets/pdb-all.parquet")
```

```python
df
```

```python
# first filter out any one with VENOM to start
def filter_out_venom(df):
	return df[df["name"].apply(lambda x: "VENOM" not in x)]
df = filter_out_venom(df)
df
```

```python
def filter_out_models(df):
	first_model = df["name"].apply(lambda x: "MODEL_1_" in x or "MODEL" not in x)
	return df[first_model]
df = filter_out_models(df)
```

```python
df[df["name"].apply(lambda x: "_C" in x)]
```

```python
def filter_out_asm(df):
	return df[df["name"].apply(lambda x: "_C" in x)]
filter_out_asm(df)
```

```python
def filter_name(df, apply_func):
	return df[df["name"].apply(apply_func)]

def filter_out_asm(df):
	return filter_name(df, lambda x: "_A" in x or "gz_" not in x)

df = filter_out_asm(df)
```

```python
df[["name", "3Di", "AA"]].to_parquet("./datasets/pdb-no-model-no-asm.parquet", index=False)
```
